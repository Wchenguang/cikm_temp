```python
def boxplot_multi_cols(dataframe, per_line = 5, figsize = (8,3), columns = None):
    if(columns == None):
        columns = dataframe.columns
    if(len(columns) % per_line == 0):
        line_num = len(columns) // per_line
    else:
        line_num = len(columns) // per_line  + 1
    for line in range(line_num):
        part_cols = columns[line*per_line : (line+1)*per_line]
        ax = plt.figure(figsize = figsize).gca()
        part_df = dataframe[part_cols]
        part_df.plot.box(ax = ax, grid = True)
        
def hist_multi_cols(dataframe, figsize = (5, 3)):
    columns = dataframe.columns
    for col in columns:
        fig = plt.figure(figsize=figsize)
        plt.hist(dataframe[col], label = col)
        plt.legend()
        plt.show()
```

* 将用户行为表的时间戳列转化为每一天的不同时段

```python
import numpy as np
'''
将时间戳转化为一天的不同时段
'''
def time_gapper(seconds, basetime = 0):
    seconds_a_day = 24 * 60 * 60
    seconds_an_hour = 60 * 60
    gaptime_a_day = (seconds - basetime) % seconds_a_day
    hours = gaptime_a_day // seconds_an_hour + gaptime_a_day % seconds_an_hour / seconds_an_hour
    res = np.full(seconds.shape, "time_a_day")
    if((hours >= 23).any()):
        res[hours >= 23] = 'mid night'
    if((hours < 7).any()):
        res[hours < 7] = 'mid night'
    if((hours>=7) & (hours < 11)).any():
        res[(hours>=7) & (hours < 11)] = 'morning'
    if((hours>=11) & (hours < 13)).any():
        res[(hours>=11) & (hours < 13)] = 'noon'
    if((hours>=13) & (hours < 18)).any():
        res[(hours>=13) & (hours < 18)] = 'afternoon'
    if((hours>=18) & (hours < 23)).any():
        res[(hours>=18) & (hours < 23)] = 'night'
    return res.reshape(-1, 1)
if(False == read_local):  
    train_behavior['interval'] = time_gapper(train_behavior['time'])
#grouped = train_behavior['interval'].groupby(train_behavior['user_id'])
```

## 分层抽样
* 抽取较少数据用于分析

```python
'''
区分不同年龄段
'''
import numpy as np
def age_gapper(frame, age_col):
    res = np.full(frame[age_col].shape, 'wealthy')
    res[frame[age_col] < 26] = "student"
    res[(frame[age_col] >= 26) & (frame[age_col] < 41)] = "adult"
    res[frame[age_col] >= 41] = "old"
    return res
'''
组合年龄与购买力属性
用于分层抽样
'''
if(False == read_local):
    train_user['age_interval'] = age_gapper(train_user, 'age')
    train_user['purchase_interval'] = train_user['purchase'] / 2 // 1

'''
分层抽样
'''
from sklearn.model_selection import StratifiedShuffleSplit
def stratified_sample(data, columns, train_size, verify_size):
    train = verify = temp = None
    split = StratifiedShuffleSplit(n_splits = 1, test_size = train_size)
    for trainIndex, testIndex in split.split(data, data[columns]):
        train = data.loc[testIndex]
        temp = data.loc[trainIndex]
    split = StratifiedShuffleSplit(n_splits = 1, test_size = verify_size / (1 - train_size))
    for trainIndex, testIndex in split.split(temp, temp[columns]):
        verify = data.loc[testIndex]
    return train, verify
if(False == read_local):
    splited_train_user, splited_verify_user = stratified_sample(
        train_user, ['sex', 'age_interval', 'purchase_interval'], 0.02, 0.005)
```

* 根据小样本用户数据，构造包含商品与行为信息的完整数据集

```python
def expand(dataframe):
    splited_behavior = train_behavior[train_behavior['user_id'].isin(dataframe['user_id'])].copy()
    splited_behavior['interval'] = time_gapper(splited_behavior['time'])
    grouped = splited_behavior['interval'].groupby(splited_behavior['user_id'])
    gc = grouped.value_counts()
    dataframe['morning'] = np.zeros((dataframe.shape[0], 1))
    dataframe['noon'] = np.zeros((dataframe.shape[0], 1))
    dataframe['afternoon'] = np.zeros((dataframe.shape[0], 1))
    dataframe['night'] = np.zeros((dataframe.shape[0], 1))
    dataframe['mid night'] = np.zeros((dataframe.shape[0], 1))
    for ziped in gc.index:
        user_id = ziped[0]
        index = dataframe[dataframe['user_id'] == user_id].index
        dataframe.loc[index, ziped[1]] = gc[ziped]
if(False == read_local):
    expand(splited_verify_user)
    expand(splited_train_user)
    splited_train_user.to_csv(splited_train_user_path)
    splited_verify_user.to_csv(splited_verify_user_path)
    splited_train_user['shopping_sum'] = splited_train_user[['morning', 'noon', 
                        'afternoon', 'night', 
                        'mid night']].sum(axis = 1)
    splited_verify_user['shopping_sum'] = splited_verify_user[['morning', 'noon', 
                        'afternoon', 'night', 
                        'mid night']].sum(axis = 1)
```

```python
def plot_nv1_scatter(data, xcols, ycol, figsize = (6,4)):
    for xcol in xcols:
        plt.figure(figsize=figsize)
        plt.scatter(x = data[xcol], y = data[ycol])
        plt.title(xcol + " vs " + ycol)
        plt.ylabel(ycol)
        plt.xlabel(xcol)
        plt.show()

plot_nv1_scatter(splited_train_user, time_cols, 'shopping_sum')
```

* 对以上数据做线性回归

```python
def plot_draw2dline_with_weight(weight, bias, 
                                title = None):
    x=np.linspace(-1, 100, 1000, endpoint=True)
    y = weight * x + bias
    plt.plot(x, y)
    if(title):
        plt.title(title)

from gglearn.ml.LinearRegression import *
linear_clf = LinearRegression()
for x_col in time_cols:
    y = linear_clf.fit(splited_train_user[[x_col]].values, splited_train_user['shopping_sum'])
    plt.scatter(x=splited_train_user[x_col], y = splited_train_user['shopping_sum'], c = 'r')
    plot_draw2dline_with_weight(linear_clf.theta[0], linear_clf.theta[1], title = x_col)
    plt.show()
    print(linear_clf.theta)
    
```
